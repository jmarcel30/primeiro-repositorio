```python
# Lógica de programação

# Passo 0 - Entender o desafio que você quer resolver.

# Passo 1 - Usar o python para percorrer todos os arquivos da pasta base de dados. (Pasta Vendas)
import os
import pandas as pd

list_arquivo = os.listdir("/content/drive/MyDrive/Curso python/Vendas/")
display(list_arquivo)

tabela_total = pd.DataFrame()

# Passo 2 - Importar a base de dados vendas


for arquivo in list_arquivo:   # para cada arquivo dentro da minha lista de arquivo: faça algo
  #se tem "vendas" no nome do arquivo, então
  if "Vendas" in arquivo:
    # importar o arquivo
      tabela = pd.read_csv(f"/content/drive/MyDrive/Curso python/Vendas/{arquivo}")
      tabela_total = tabela_total.append(tabela)





# Passo 3 - Tratar/ Compilar as bases de dados
display(tabela_total)


# Passo 5 - Calcular o produto que mais faturou (em faturamento)

# Passo 6 - Calcular a loja/Cidade que mais vendeu (em faturamento) - criar um gráfico/dasboard
```

```python
from sqlalchemy import true
# Passo 4 - Calcular o produto mais vendido (em quantidade)
tabela_produto = tabela_total.groupby('Produto').sum()[['Quantidade Vendida']].sort_values(by='Quantidade Vendida' ,ascending=False)
display(tabela_produto)
```
